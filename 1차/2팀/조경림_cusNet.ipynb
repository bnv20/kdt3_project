## **반기별/품목별 증감액** 

```python
반기14_1 = data.query('년==2014&분기==1|분기==2').groupby('분류 체계')['구매금액'].sum().to_frame()
반기14_2 = data.query('년==2014&분기==3|분기==4').groupby('분류 체계')['구매금액'].sum().to_frame()
반기15_1 = data.query('년==2015&분기==1|분기==2').groupby('분류 체계')['구매금액'].sum().to_frame()
반기15_2 = data.query('년==2015&분기==3|분기==4').groupby('분류 체계')['구매금액'].sum().to_frame()


반기14_1.rename(columns={'구매금액':'2014_1분기_구매금액'}, inplace=True)
반기14_2.rename(columns={'구매금액':'2014_2분기_구매금액'}, inplace=True)
반기15_1.rename(columns={'구매금액':'2015_1분기_구매금액'}, inplace=True)
반기15_2.rename(columns={'구매금액':'2015_2분기_구매금액'}, inplace=True)
```

```python
반기2014_12 = pd.concat([반기14_1,반기14_2], axis=1)
반기2014_2_2015_1 = pd.concat([반기14_2,반기15_1], axis=1)
반기2015_12 = pd.concat([반기15_1,반기15_2], axis=1)
```

```python
증감_1 = (반기14_2['2014_2분기_구매금액'] - 반기14_1['2014_1분기_구매금액'])/반기14_1['2014_1분기_구매금액'] *100
증감_2 = (반기15_1['2015_1분기_구매금액'] - 반기14_2['2014_2분기_구매금액'])/반기14_2['2014_2분기_구매금액'] *100
증감_3 = (반기15_2['2015_2분기_구매금액'] - 반기15_1['2015_1분기_구매금액'])/반기15_1['2015_1분기_구매금액'] *100

증감_1 = 증감_1.to_frame()
증감_2 = 증감_2.to_frame()
증감_3 = 증감_3.to_frame()

증감_1.rename(columns={0 : '증감%_1'}, inplace=True)
증감_2.rename(columns={0 : '증감%_2'}, inplace=True)
증감_3.rename(columns={0 : '증감%_3'}, inplace=True)
```

```python
df1 = pd.concat([반기2014_12, 증감_1], axis=1)
df2 = pd.concat([반기2014_2_2015_1, 증감_2], axis=1)
df3 = pd.concat([반기2015_12, 증감_3], axis=1)

display(df1, df2, df3)
```

```python
df = data.groupby(['고객번호','년','분기'])['구매금액'].agg(['sum', 'count'])
df1 = df.reset_index(2).query('분기==1|분기==2')

df1 = df1.reset_index()
df1 = df1.groupby(['고객번호','년']).sum()
df1.reset_index()
```

```python
df1.reset_index(inplace=True)
```

```python
df2 = pd.pivot_table(df1,
               columns='년',
               index='고객번호',
               values='sum')
```

```python
df2.dropna(axis=0,inplace=True)
display(df2, df2.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df2_scaled = scaler.fit_transform(df2)
df2_scaled
```
