```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dataset/data.csv')
data.head()
```

```python
# 상품별 인기 TOP 100

인기상품 = data['소분류명'].value_counts()[:100]
인기상품 = pd.DataFrame(data=인기상품)
인기상품.rename(columns = {'소분류명' : '총 판매갯수'}, inplace = True)
인기상품.head(10)
```

```python
인기상품2 = 인기상품.reset_index()
인기상품2.rename(columns = {'index' : '상품명'}, inplace = True)
인기상품2.head()
```

```python
# 전체 상품별 인기순위 TOP 10

x = 인기상품2['상품명'][:10]
y = 인기상품2['총 판매갯수'][:10]
plt.bar(x, y)
plt.xticks(rotation = 45)
plt.title('전체 상품별 인기순위 TOP 10', fontsize=15)
plt.show()
```

```python
# 소,중,대분류명 별로 인기상품 순위 출력이 필요한가
```

```python
# 연령대는 총 10개의 범위로 나누어져 있음
# 각 연령대별로 파이차트 시각화하여 소분류명의 상품 갯수의 분포 파악 필요함
A = data.query('연령대=="19세이하"')
B = data.query('연령대=="20세~24세"')
C = data.query('연령대=="25세~29세"')
D = data.query('연령대=="30세~34세"')
E = data.query('연령대=="35세~39세"')
F = data.query('연령대=="40세~44세"')
G = data.query('연령대=="45세~49세"')
H = data.query('연령대=="50세~54세"')
I = data.query('연령대=="55세~59세"')
J = data.query('연령대=="60세이상"')

age = [A,B,C,D,E,F,G,H,I,J]
```

```python
# 전체기간(14,15년도) 연령대별 인기상품 TOP 10
fig, axs = plt.subplots(figsize=(20,15), nrows=5, ncols=2)
# plt.subplots(constrained_layout=True) # 서브플롯간의 간격을 가능한 최적의 수치로 자동 조정
plt.subplots_adjust(left=0.125, bottom=2, right=0.9, top=4, hspace=0.7,wspace=0.2)


for num,i in enumerate(age):
  row = int(num/2); col = num%2
  i.groupby('연령대')['소분류명'].value_counts()[:10].plot(kind='bar', ax=axs[row][col], title='{} 인기상품 TOP 10'.format(i['연령대'].unique()[0]))
```

```python
# 상품 간 연관성을 어떻게 추측할 것인가.
# 소분류명 별로 상품을 분류했을때, 효율성이 떨어짐 => 중분류명으로 분류해볼것
# 연령대별로 분기마다 구매비용, 구매횟수 분석필요
```

```python
first = data.query('분기==1')
second = data.query('분기==2')
third = data.query('분기==3')
four = data.query('분기==4')

seasons = [first, second, third, four]
```

```python
data.query('연령대=="19세이하"&분기==1').groupby('중분류명')['구매금액'].sum().sort_values(ascending=False)[:10].plot(kind='bar')
plt.title('19세 이하 1분기 중분류 상품 TOP 10')
```

```python
# 연령대별 각 분기의 총 구매금액
fig, axs = plt.subplots(figsize=(10,25), nrows=5, ncols=2)
plt.subplots_adjust(hspace=0.7,wspace=0.2)

for num, i in enumerate(age):
  row = int(num/2); col = num%2
  i.groupby('분기')['구매금액'].sum().plot(kind='bar', ax=axs[row][col], title='{} 분기별 구매금액'.format(i['연령대'].unique()[0]))
```

```python
# 연령대별 각 분기의 총 구매횟수
fig, axs = plt.subplots(figsize=(10,25), nrows=5, ncols=2)
plt.subplots_adjust(hspace=0.7,wspace=0.2)

for num, i in enumerate(age):
  row = int(num/2); col = num%2
  i.groupby('분기')['영수증번호'].count().plot(kind='bar', ax=axs[row][col], title='{} 분기별 구매횟수'.format(i['연령대'].unique()[0]))
```
