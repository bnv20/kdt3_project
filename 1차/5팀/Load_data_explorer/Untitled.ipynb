```python
df_A = df[df['제휴사']=='A']
```

```python
def get_month(구매일자):    # 2014년도 월은 1~12   2015년도 월은 13~24
    x=str(구매일자)
    month=int(x[4:6])
    return month

def get_year(구매일자):    # 2014년도 월은 1~12   2015년도 월은 13~24
    x=str(구매일자)
    yr = int(x[:4])
    return yr
```

```python
#롯데 백화점
df_A['구매연도'] = df_A['구매일자'].apply(get_year)
```

```python
df_A['구매월'] = df_A['구매일자'].apply(get_month)
```

```python
df_A = df_A.drop('구매일자',axis=1)
```

```python
df_A = df_A.sort_values(by=['구매연도','구매월'])
```
