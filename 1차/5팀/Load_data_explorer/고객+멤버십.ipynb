```python
import pandas as pd
df_cust = pd.read_csv('../LOTT/고객DEMO.txt',sep=',',encoding='cp949')
df_cust
```

```python
df_mem = pd.read_csv('../LOTT/멤버십여부.txt',sep=',',encoding='cp949')
```

```python
data_mem = df_mem.groupby('고객번호')[['멤버십명']].count()
```

```python
data_mem = data_mem.reset_index()
```

```python
data_mem
```

```python
data_mem
```

```python
df_mem.info()
```

```python
data_mem = data_mem.rename(columns={'멤버십명':'멤버십가입개수'})
```

```python
cust = pd.merge(df_cust,data_mem,on='고객번호',how='left')
```

```python
cust['멤버십가입개수'] = cust['멤버십가입개수'].fillna(0).astype(int)
```
