```python
def get_year(구매일자):    # 2014년도 월은 1~12   2015년도 월은 13~24
    x=str(구매일자)
    month=int(x[4:6])
    yr = int(x[:4])
    return yr
```

```python
Total['구매년도'] = Total['구매일자'].apply(get_year)
```

```python
t_amount = Total.groupby('구매년도')[['구매금액']].sum()
```

```python
t_amount = t_amount.reset_index()
```

```python
t_amount.to_csv('../dataset/연도별전체매출액.csv')
```
