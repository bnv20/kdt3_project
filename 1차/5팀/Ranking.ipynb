```python
Li = ['가공식품', '가구/인테리어', '교육/문화용품', '기타', '디지털/가전', '명품', '신선식품', '의류',
       '일상용품', '전문스포츠/레저', '패션잡화']
```

```python
def ratio(df):
  test = df[['고객번호','대분류명']]
  test = test.set_index('고객번호')
  A = pd.get_dummies(test['대분류명']).reset_index()
  A = A.groupby('고객번호').sum().astype(int)
  A['total'] = A.sum(axis=1).astype(int)
  for i in Li:
    A[f"{i}_r"] = A[i] / A['total']
  B = A.drop(['가공식품', '가구/인테리어', '교육/문화용품', '기타', '디지털/가전', '명품', '신선식품', '의류',
    '일상용품', '전문스포츠/레저', '패션잡화','total'],axis=1)
  B['total'] = B.sum(axis=1)
  return B
```

```python
H1_r = ratio(df_1)
H2_r = ratio(df_2)
H3_r = ratio(df_3)
H4_r = ratio(df_4)
```

```python
def Makingstandard(df):
  df = df.reset_index(drop=False)
  df = df.rename(columns={'index':'고객번호'})
  df = df.set_index('고객번호',drop=True)
  return df

pr_1 = Makingstandard(pr_1)
pr_2 = Makingstandard(pr_2)
pr_3 = Makingstandard(pr_3)
pr_4 = Makingstandard(pr_4)
```

```python
pr_1 = pr_1[sorting_cols]
pr_2 = pr_2[sorting_cols]
pr_3 = pr_3[sorting_cols]
pr_4 = pr_4[sorting_cols]
```

```python
def ratio(df):
  global sorting_cols, def_cols, sum_cols, confirm_cols
  s_df = df[sorting_cols]
  for i in def_cols:
    s_df[f'{i}_r'] = s_df[i] / s_df['총구매금액']
  s_df['total'] = s_df[sum_cols].sum(axis=1)
  s_df = s_df[confirm_cols]
  return s_df
```
